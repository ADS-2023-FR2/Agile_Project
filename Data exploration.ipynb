```python
import matplotlib.pyplot as plt
from spotlight.datasets.movielens import get_movielens_dataset

# Load the MovieLens 100K dataset
dataset = get_movielens_dataset(variant='100K')

# Access the user-item interaction data
interactions = dataset

# Plot the distribution of ratings
ratings = interactions.ratings
```

```python
type(dataset).__name__
```

```python
plt.hist(ratings, bins=5, range=(1, 5), alpha=0.7)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
# Analyze the number of ratings per movie without 'groupby'
item_counts = {}
for item_id in interactions.item_ids:
    item_counts[item_id] = item_counts.get(item_id, 0) + 1

ratings_per_movie = list(item_counts.values())
plt.hist(ratings_per_movie, bins=30, alpha=0.7)
plt.title('Number of Ratings Per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Frequency')
plt.show()
```
